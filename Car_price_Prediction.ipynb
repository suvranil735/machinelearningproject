```python
df1 = df[['symboling','fueltype','doornumber','enginetype','enginesize','horsepower','compressionratio','highwaympg','price']]
```

```python
df.columns
```

```python
df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df1['fueltype']=le.fit_transform(df1['fueltype'])
```

```python
df1.head(2)
```

```python
df1['doornumber']=le.fit_transform(df1['doornumber'])
df1['enginetype']=le.fit_transform(df1['enginetype'])
```

```python
df1.head(2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
cols=['enginesize','horsepower','highwaympg']
```

```python
df1[cols]=ss.fit_transform(df1[cols])
```

```python
df1.head(2)
```

```python
#df1['enginesize']=ss.fit_transform(df1[['enginesize']])
```

```python
from sklearn.model_selection import train_test_split
```

```python
df1.columns
```

```python
X = df1[['symboling', 'fueltype', 'doornumber', 'enginetype', 'enginesize','horsepower', 'compressionratio', 'highwaympg']]
Y = df1['price']
```

```python
X.shape
```

```python
X_train, X_test, Y_train,Y_test = train_test_split(X, Y, test_size=0.3)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt = DecisionTreeRegressor()
```

```python
dt.fit(X_train,Y_train)
```

```python
y_pred = dt.predict(X_test)
```

```python
y_pred
```

```python
Y_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mean_absolute_error(Y_test,y_pred)
```

```python
mean_squared_error(Y_test,y_pred)
```

```python
dt.feature_importances_
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor()
```

```python
rf.fit(X_train,Y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
mean_absolute_error(Y_test,y_pred_rf)
```

```python
mean_squared_error(Y_test,y_pred_rf)
```
